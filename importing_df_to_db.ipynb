```python
def connect_to_db(host_name, dbname, port, username, password):
    try:
        conn = ps.connect(host=host_name, database=dbname, user=username, password=password, port=port)

    except ps.OperationalError as e:
        raise e
    else:
        print('Connected!')
        return conn

def create_table(curr):
    create_table_command = ("""CREATE TABLE IF NOT EXISTS songs (
                    
                    artist VARCHAR(255) NOT NULL,
                    album VARCHAR(255) NOT NULL,
                    track_name VARCHAR(255) ,
                    popularity INTEGER NOT NULL,
                    img_url VARCHAR(255) ,
                    danceability FLOAT  NOT NULL,
                    energy FLOAT NOT NULL,
                    key INTEGER NOT NULL,
                    loudness FLOAT NOT NULL,
                    mode INTEGER NOT NULL,
                    speechiness FLOAT NOT NULL,
                    instrumentalness FLOAT ,
                    liveness FLOAT NOT NULL,
                    valence FLOAT NOT NULL,
                    tempo FLOAT NOT NULL,
                    text_data VARCHAR(255) NOT NULL
                    )""")
            
                    
                    
            

    curr.execute(create_table_command)

def insert_into_table(curr,artist,album,track_name,popularity,img_url,danceability,energy,key,loudness,
                      mode,speechiness,instrumentalness,liveness,valence,tempo,text_data):
    
    insert_into_videos = ("""INSERT INTO songs (artist,album,track_name,popularity,img_url,danceability,energy,key,loudness,
                      mode,speechiness,instrumentalness,liveness,valence,tempo,text_data)
    VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);""")
    
    row_to_insert = (artist,album,track_name,popularity,img_url,danceability,energy,key,loudness,
                      mode,speechiness,instrumentalness,liveness,valence,tempo,text_data)
    curr.execute(insert_into_videos, row_to_insert)


def append_from_df_to_db(curr,df):
    for i, row in df.iterrows():
        insert_into_table(curr, row['artist'], row['album'], row['track_name'], row['popularity'],row["img_url"]
                          , row['danceability'], row['energy'], row['key'], row['loudness'], row['mode'], row['speechiness'], 
                          row['instrumentalness'], row['liveness'] , row['valence'], row['tempo'],row['text_data'])

                         

```
